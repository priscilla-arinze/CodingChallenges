# LeetCode: [Group Anagrams](https://leetcode.com/problems/group-anagrams/)

__Difficulty: Medium__

__Data Structure(s): String, Array, Dictionary/Hashmap__

__Technique(s): ASCII math, 26 zeros list for alphabet__

<hr>

### Description:

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

 

Example 1:

> Input: `strs = ["eat","tea","tan","ate","nat","bat"]`<br>Output: `[["bat"],["nat","tan"],["ate","eat","tea"]]`

Example 2:

> Input: strs = `[""]`<br>Output: `[[""]]`

Example 3:

> Input: strs = `["a"]`<br>Output: `[["a"]]`


 

Constraints:

* 1 <= strs.length <= 104
* 0 <= strs[i].length <= 100
* strs[i] consists of lowercase English letters.

 

<hr>

Time Complexity: $O(K * N)$
* $K$: average number of characters in each string (for the sake of counting character frequencies)
* $N$: number of input strings in list

```python
def groupAnagrams(input_strs):
    
    # Key: character frequency count (tuple)
    # Value: list of valid anagrams strings
    dict_CharFreqTupleToAnagramStrList = {}


    for string in input_strs:
        # an array of 26 zeros; each index mapping to an alphabet letter's frequency count
        # will be updated for EACH input string
        list_AtoZ_CharFreqMappings = [0] * 26 

        
        for char in string:
            # e.g. ASCII math: 'a'(97) - 'a'(97) = index 0;  'z'(122) - 'a'(97) = index 25
            list_AtoZ_CharFreqMappings[ord(char) - ord("a")] += 1  # increment the frequency of each chracter in string 


        # at the end of counting each character frequency for each individual string:
            # convert AtoZ_CharFreqMappings to tuple
            # KEY -- tuple of char counts, VALUE -- string in current iteration
        tuple_AtoZ_CharFreqMappings = tuple(list_AtoZ_CharFreqMappings)

        # IF: tuple key already exists in dictionary, APPEND to anagrams list
        if tuple_AtoZ_CharFreqMappings in dict_CharFreqTupleToAnagramStrList:
            dict_CharFreqTupleToAnagramStrList[tuple_AtoZ_CharFreqMappings].append(string)

        # ELSE: set new key equal to list type containing the current string in iteration
        else:
            dict_CharFreqTupleToAnagramStrList[tuple_AtoZ_CharFreqMappings] = [string]


    return list(dict_CharFreqTupleToAnagramStrList.values()) # anagrams list
```

```python
strs = ["eat","tea","tan","ate","nat","bat"]

print(groupAnagrams(strs))
```
