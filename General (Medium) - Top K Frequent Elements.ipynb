# LeetCode: [Top K Frequent Elements](https://leetcode.com/problems/top-k-frequent-elements)

__Difficulty: Medium__

__Data Structure(s): Array, Hashmap/Dictionary__

__Techniques(s): Variation of bucket sort (counts as indexes), iterating backwards with while loop__

<hr>

### Description:

Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

 

Example 1:

> Input: nums = [1,1,1,2,2,3], k = 2<br>Output: [1,2]


Example 2:

> Input: nums = [1], k = 1<br>Output: [1]


Constraints:
* `1 <= nums.length <= 105`
* `-104 <= nums[i] <= 104`
* `k` is in the range [1, the number of unique elements in the array].
* It is __guaranteed__ that the answer is __unique__.




<hr>

Time Complexity: $O(N)$

Space Complexity: $O(N + N) = O(N)$

```python
nums = [1, 3, 450, 1, 1, 450, 450, 2, 2, 450]
k = 2

numsCounts = dict()
numsFreqList = []
result = []

# create an array of N arrays
for i in range(len(nums) + 1):
    numsFreqList.append([])

# map counts
for num in nums:
    numsCounts[num] = numsCounts.get(num, 0) + 1

# append keys for applicable value index (e.g. 1, freq: 3 --> 1 appended to 3rd index)
for key, val in numsCounts.items():
    numsFreqList[val].append(key)
    print(key, val)

# iterate backwards
rightPointer = len(numsFreqList) - 1

while rightPointer >= 0:
    if (len(result) == k):
        break

    if (len(numsFreqList[rightPointer]) > 0):
        for n in numsFreqList[rightPointer]:
            if (len(result) == k):
                break
            result.append(n)

    rightPointer -= 1

print(numsFreqList)
print(result)
```
