# LeetCode: [Valid Parentheses](https://leetcode.com/problems/valid-parentheses/description/)

__Difficulty: Easy__

__Data Structure(s): Stack, Dictionary/HashMap__

<hr>

### Description:

Given a string s containing just the characters `'(', ')', '{', '}', '[' and ']'`, determine if the input string is valid.

An input string is valid if:
* Open brackets must be closed by the same type of brackets.
* __Open brackets must be closed in the correct order.__
* Every close bracket has a corresponding open bracket of the same type.




 

Example 1:

> Input: s = "()"<br>Output: true

Example 2:

> Input: s = "()[]{}"<br>Output: true

Example 3:

> Input: s = "(]"<br>Output: false

Example 4:
> Input: s = "([)]"<br>Output: false

Example 5:
> Input: s = "{[]}"<br>Output: true

Constraints:
* 1 <= s.length <= 104
* s consists of parentheses only '()[]{}'.


<hr>

*Initial Attempt (using a stack)*

Time Complexity: $O(N)$

Space Complexity: $O(N)$ -- *the size of the stack will be dependent on the size of the string*

```python
def isValidParentheses(s):
    """
    :type s: str
    :rtype: bool
    """
    stack = []
    closeToOpenPairs = {
        ')': '(',
        '}': '{',
        ']': '['
    }
    
    for char in s:
        # check to see if the current character is one of the closing type (dictionary keys)
        if char in closeToOpenPairs:
            # If closing type, make sure that the stack is not empty 
                # AND make sure that the very last character is the matching opening type (dictionary value) of the current closing type
            # If true, remove/pop that matching opening type from the stack
            if len(stack) != 0 and stack[-1] == closeToOpenPairs[char]:
                stack.pop()
            else:
                return False
        
        # push/append opening type only
        else:  
            stack.append(char)

    # FINALLY, check to see if the stack has been emptied out 
        # in other words, all opening types should have been accounted for & popped out of stack
    return True if len(stack) == 0 else False

print(isValidParentheses("()"))
print(isValidParentheses("()[]{}"))
print(isValidParentheses("(]"))
print(isValidParentheses("([)]"))
print(isValidParentheses("{[]}"))
```
